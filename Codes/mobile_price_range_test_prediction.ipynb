#  Prediction of Mobile Price Range on Test dataset Model 

```python
import pandas as pd
import joblib
import warnings
warnings.filterwarnings('ignore')
```

```python
df_test = pd.read_csv("../Datasets/test.csv")
```

```python
df_test.head()
```

```python
df_test.shape
```

```python
features = joblib.load('../model/features.h5')
features
```

```python
x = df_test[features]
x
```

```python
## load model, scalar, pca
```

```python
scaler = joblib.load('../model/scaler.h5')
pca = joblib.load('../model/pca.h5')
model = joblib.load('../model/model.h5')
```

```python
model
```

```python
scaled_x = scaler.transform(x)
scaled_x
```

```python
scaled_x = pca.transform(scaled_x)
```

```python
predict_price_range = model.predict(scaled_x)
```

```python
predict_price_range
```

#### add prediction on the df_prediction

```python
df_prediction = df_test[features]
df_prediction['id'] = df_test['id']
df_prediction.head()
```

```python
df_prediction['predicted_price_range'] = predict_price_range
df_prediction.head()
```

```python
df_prediction.tail()
```

#### save prediction on new csv file 

```python
predict_test = df_prediction[['id','predicted_price_range']]
predict_test.head()
```

```python
predict_test.to_csv('../Datasets\predict_test.csv',index=False)
```
